Importing the Library 'matplotlib.pyplot' used as a shortcut 'plt'. 


Using 'plot' command to plot the points in the graph - '[1,2,3,4,5],[2,4,9,16,25]'.
 - 'go-' where g is for green, o is for dot and - is for line connecting dots.Similar examples can be found in the link -  '  https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot '
 - 'label' is to define the name of the line.
 - 'axis' is to define the length of the graph 0,6 for x and 0,10 for Y.
 - 'legend' for the corner box defining line 1 and line 2 with colors.
 - 'show' is to show the graph

matplotlib.pyplot.axis(*v,**kwargs).

```python
import matplotlib.pyplot as plt 
plt.plot([1,2,3,4,5],[2,4,9,16,25],'go-', label = 'line 1',linewidth=2)
plt.plot([2,3,4],[4,9,16],'ro-', label = 'line 2')
plt.axis([0,6,0,10],'image')
plt.legend()
plt.ylabel('Y')
plt.xlabel('x')
plt.show()
```

 
Importing 'numpy' as np to play with numbers more conveniently



t holds all the numbers designed with (0, 5 , 0.2)
  - 0,5 is the range and 0.2 is the interval 

Using 'plot' command to plot the 't' in different forms 

```python
import numpy as np 

# evenly sampled time at 0.2 intervals 
t = np.arange(0, 5, 0.2)

# representing numbers with colors
plt.plot(t, t, 'ro-', t, t**2, 'bo-', t ,t**3,'go-')
plt.show()
```

Importing Numpy and Matplotlib.pyplot as np and plt to play with numbers and plot them in graph.


subplot(nrows, ncols, plot_number)
 - here it is subplot(211) 

figure is set to 1 otherwise figure(1) will be created by default.

```python
import numpy as np 
import matplotlib.pyplot as plt 

def f(t):
    return np.exp(-t) * np.cos(2 * np.pi*t)

t1 = np.arange(0.0, 100.0, 5)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2),'r--')
plt.show()
```

Creaing multiple figures 
   plt.figure(1)
   plt.figure(2)


Current figure can be clear with clf() and axes with cla()

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

If you are making lots of figures, you need to be aware of one more thing: the memory required for a figure is not completely released until the figure is explicitly closed with close().


Deleting all references to the figure, and/or using the window manager to kill the window in which the figure appears on the screen, is not enough, because pyplot maintains internal references until close() is called.

matplotlib.pyplot.close(*args)

# Working with texts
  - Adding text in the arbitrary location with text() command. 
  - plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
  



Adding text in the bars or giving title of the graph.   
- plt.xlabel('Smarts')
- it can be further customized with - fontsize=14, color='red'  
- plt.ylabel('Probability')
- plt.title('Histogram of IQ') 

plt.grid(True) 
  - To plot grid lines on the graph.

```python
import numpy as np 
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts',fontsize=14, color='red' )
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

plt.plot(t,s)
plt.title(r'$\alpha_i > \beta_i$', fontsize=20) 
plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=30)
plt.text(0.6, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',
         fontsize=20)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
```

A common use for text is to annotate some feature of the plot, and the annotate() method provides helper functionality to make annotations easy.


  

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),)
  - 'local max' is the text with what we want to annotate. 
  -  xy=(2, 1) is the position of the point to annotate.
  -  xytext=(3, 1.5) is the position of the text.
  -  arrowprops=dict(facecolor='black', shrink=0.05) defining the arrow color and shrinkage in the graph.

plt.ylim(-2,2) is to define the range of y from -2 to 2 in the graph.


```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),)

plt.ylim(-2,2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.ticker import NullFormatter  # useful for `logit` scale

# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(1)

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
